Korelasi antar variabel

```python
def strokes_corr(strokes):
  correlation = strokes.corr()
  sns.heatmap(correlation, annot=True, cbar=True, cmap="RdYlGn")
strokes_corr(strokes)
```

```python
strokes.corr()
```

```python
x = strokes.iloc[:,:-1]
y = strokes.iloc[:,-1]
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(), [0,4,5,6,9])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
```

Splitting Dataset into Training & Testing

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=1)
```

Scalling Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train_1 = sc.fit_transform(x_train)
```

```python
x_test_1 = sc.fit_transform(x_test)
```

Using SVM

```python
from sklearn.svm import SVC
```

```python
classifier = SVC(random_state = 0, kernel = 'linear')
```

```python
classifier.fit(x_train_1, y_train)
```

```python
y_pred = classifier.predict(x_test_1)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

HASIL AKURASI

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(y_test, y_pred)
acc
```

```python
from sklearn.metrics import classification_report
```

```python
print(f"Report Model Evaluation : \n {classification_report(y_test, y_pred)}")
```
